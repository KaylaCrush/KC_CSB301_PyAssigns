Some hints for this assignment:    
Your goal is to create conditions to ensure that the expressions that you pass into the function are exactly what is needed. For modus ponens, you need exactly premise1 = Implies(p, q) and premise2 = p. Be careful here, you cannot use a condition like if premise1 and premise2:, becuase then any premises will be accepted. Look up the Python isinstance() function similar to what is shown in the Modus Ponens function below.
```python
# Inferences

from sympy.logic.boolalg import Or, Not, And, Implies
from sympy.abc import p, q, r

def modus_ponens(premise1, premise2):
    """Modus Ponens: If p implies q, and p is true, then q is true."""
    if premise2 and isinstance(premise1, Implies):
        if premise1.args[0] == premise2:
            return premise1.args[1]

def modus_tollens(premise1, premise2):
    """Modus Tollens: If p implies q, and q is false, then p is false."""
    if isinstance(premise1, Implies):
        if premise1.args[1] == ~premise2:
            return ~premise1.args[0]

def hypothetical_syllogism(premise1, premise2):
    """Hypothetical Syllogism: If p implies q, and q implies r, then p implies r."""
    if isinstance(premise1, Implies) and isinstance(premise2, Implies) and premise1.args[1] == premise2.args[0]:
        return Implies(premise1.args[0], premise2.args[1])

def disjunctive_syllogism(premise1, premise2):
    """Disjunctive Syllogism: If p or q is true, and p is false, then q is true."""
    if isinstance(premise1, Or) and premise1.args[0] == ~premise2:
        return premise1.args[1]


def addition(premise1, premise2):
    """Addition: If p is true, then p or q is true."""
    return Or(premise1, premise2)

def conjunction(premise1, premise2):
    """Conjunction: If p is true, and q is true, then p and q are true"""
    return premise1 & premise2


def simplification(premise):
    """Simplification: If p and q is true, then p is true."""
    if isinstance(premise, And):
        return premise.args[0]


def resolution(premise1, premise2):
    """Resolution: If (p or q) and (not p or r) are true, then q or r is true."""
    if isinstance(premise1, Or) and isinstance(premise2, Or):
        if premise1.args[0] == ~premise2.args[1]:
            return premise1.args[1] | premise2.args[0]

# Example premises
premise1 = Implies(p, q)
premise2 = Implies(q, r)

# Modus Ponens
conclusion_mp = modus_ponens(premise1, p)
print("Modus Ponens: If p implies q, and p is true, then q is true.")
print("Premise 1 (p implies q):", premise1)
print("Premise 2 (p is true):", p)
print("Conclusion (q is true):", conclusion_mp)
print()

# Modus Tollens
conclusion_mt = modus_tollens(premise1, ~q)
print("Modus Tollens: If p implies q, and q is false, then p is false.")
print("Premise 1 (p implies q):", premise1)
print("Premise 2 (q is false):", ~q)
print("Conclusion (q is true):", conclusion_mt)
print()


# Hypothetical Syllogism
conclusion_hs = hypothetical_syllogism(premise1, premise2)
print("Hypothetical Syllogism: If p implies q, and q implies r, p implies r.")
print("Premise 1 (p implies q):", premise1)
print("Premise 2 (q implies r):", premise2)
print("Conclusion (p implies r):", conclusion_hs)
print()

# Disjunctive Syllogism
disj_premise1 = p|q
disj_premise2 = ~p
conclusion_ds = disjunctive_syllogism(disj_premise1, disj_premise2)
print("Disjunctive Syllogism: If p or q, and not p, then q is true.")
print("Premise 1 (p or q):", disj_premise1)
print("Premise 2 (not p):", disj_premise2)
print("Conclusion (q is true):", conclusion_ds)
print()

# Addition
add_premise1 = p
add_premise2 = q
conclusion_add = addition(add_premise1, add_premise2)
print("Addition: If p is true, then p or q is true")
print("Premise 1 (p):", add_premise1)
print("Conclusion (p or q):", conclusion_add)
print()

# Conjunction
conj_premise1 = p
conj_premise2 = q
conclusion_conj = conjunction(conj_premise1, conj_premise2)
print("Conjunction: if p is true, and q is true, then p and q are true")
print("Premise 1 (p): ", conj_premise1)
print("Premise 2 (q): ", conj_premise2)
print("Conclusion (p and q): ", conclusion_conj)
print()


# Simplification
simp_premise =  p & q
conclusion_simp = simplification(simp_premise)
print("Simplification: If p and q are true, then p is true")
print("Premise (p and q):", simp_premise)
print("Conclusion (q is true):", conclusion_simp)
print()


# Resolution
res_premise1 = p | q
res_premise2 = ~p | r
conclusion_res = resolution(res_premise1, res_premise2)
print("Resolution: If (p or q) and (not p or r) are true, then q or r is true.")
print("Premise 1 (p or q): ", res_premise1)
print("Premise 2 (~p or r): ", res_premise2)
print("Conclusion (q or r): ", conclusion_res)
```
